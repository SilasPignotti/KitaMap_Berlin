# Kita-Kapazitätsanalyse Berlin - Datenaufbereitung und -schätzung

Dieses Notebook ist Teil des Projekts "KitaVision Berlin" und dient der Aufbereitung und Analyse von Kindertagesstätten-Daten aus OpenStreetMap (OSM). Das zentrale Ziel ist die Erstellung eines vollständigen Datensatzes der Berliner Kitas mit ihren Kapazitäten, der als Grundlage für weitere Versorgungsanalysen dient.

## Kernaufgaben des Notebooks:

1. **Datenintegration**: 
  - Einlesen der OSM-Kita-Daten und Berliner Bezirksgrenzen
  - Räumliche Zuordnung der Kitas zu Bezirken

2. **Kapazitätsschätzung**:
  - Analyse bekannter Kapazitätswerte
  - Entwicklung verschiedener Schätzmodelle für fehlende Kapazitäten:
    - Flächenbasierte Regression für Polygon-Geometrien
    - Bezirksspezifische Medianschätzung für Punkt-Geometrien
  - Qualitätssicherung durch Ausreißererkennung und -behandlung

3. **Datenvalidierung**:
  - Skalierung der Schätzungen auf bekannte Gesamtkapazität
  - Statistische Analysen und Visualisierungen zur Plausibilitätsprüfung

Der resultierende Datensatz bildet die Grundlage für die Analyse der räumlichen Verteilung von Kita-Plätzen und die Identifikation möglicher Versorgungslücken im Berliner Stadtgebiet.

```python
# Import necessary libraries
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from pathlib import Path

# Define constants
GEOJSON_PATH = Path("../data/raw/kitas_osm.geojson")  # Path to the raw Kita data in GeoJSON format
WFS_URL = "https://gdi.berlin.de/services/wfs/alkis_bezirke?REQUEST=GetCapabilities&SERVICE=wfs"  # URL for the Web Feature Service (WFS) for Berlin districts
OUTPUT_FILE = Path("../data/processed/kitas_processed.geojson")  # Path for the processed Kita data output
PROJ_CRS = "EPSG:32633"  # Coordinate Reference System (CRS) for UTM Zone 33N, suitable for Germany
TARGET_CAPACITY = 200_000  # Target capacity for Berlin
```

## Laden der Datengrundlagen für die KitaVision-Analyse

In diesem ersten Schritt laden wir zwei zentrale Datensätze, die die Grundlage für unsere Analyse der Kindergartenversorgung in Berlin bilden:

* **Kindertagesstätten in Berlin**: Aus OpenStreetMap beziehen wir detaillierte Standortdaten aller erfassten Kindertagesstätten im Berliner Stadtgebiet. Diese Daten liegen im GeoJSON-Format vor und enthalten neben den geografischen Koordinaten auch wichtige Attribute wie Namen und ggf. Kapazitäten der Einrichtungen.
* **Berliner Bezirksgrenzen**: Über einen Web Feature Service (WFS) greifen wir auf die offiziellen Verwaltungsgrenzen der Berliner Bezirke zu. Diese Geodaten ermöglichen uns später eine bezirksspezifische Analyse der Kitaversorgung.

Die ersten Zeilen beider Datensätze werden angezeigt, um einen ersten Überblick über die verfügbaren Informationen zu gewinnen und die Datenqualität zu prüfen.



```python
# Load Kita data from GeoJSON file
gdf = gpd.read_file(GEOJSON_PATH)
# Load Berlin districts data from Web Feature Service (WFS) URL
berlin_districts = gpd.read_file(WFS_URL)

# Display an overview of the data
print("\nKita Data:")
display(gdf.head())
print("\nDistrict Data:")
display(berlin_districts.head())
```

## Datenaufbereitung und erste räumliche Visualisierung

In dieser Phase bereiten wir die Rohdaten für die weitere Analyse auf und erstellen eine erste kartografische Übersicht:

1. **Vereinheitlichung der geografischen Bezugssysteme**:
Die Daten werden in das UTM-Koordinatensystem (Zone 33N) transformiert. Dies ermöglicht präzise Distanzberechnungen in Metern, was für spätere Erreichbarkeitsanalysen essentiell ist.

2. **Datenbereinigung**:
- Fehlende Werte werden einheitlich als NaN gekennzeichnet
- Bei Kitas ohne Namen wird "unbekannt" eingetragen
- Die Kapazitätswerte werden in ein numerisches Format konvertiert

3. **Räumliche Zuordnung**:
Durch eine räumliche Verschneidung (Spatial Join) wird jeder Kita der entsprechende Berliner Bezirk zugeordnet. Dies ermöglicht später bezirksspezifische Auswertungen der Versorgungssituation.

4. **Kartendarstellung**:
Die abschließende Visualisierung zeigt die räumliche Verteilung aller Kita-Standorte (rote Punkte) über die Berliner Bezirke. Diese Übersichtskarte gibt einen ersten Eindruck von der räumlichen Verteilung der Kinderbetreuungseinrichtungen im Stadtgebiet und lässt bereits Unterschiede in der Versorgungsdichte erkennen.

```python
# Transform coordinate reference system (CRS) to UTM Zone 33N
if gdf.crs.is_geographic:
    gdf = gdf.to_crs(PROJ_CRS)
berlin_districts = berlin_districts.to_crs(gdf.crs)

# Basic data cleaning: replace None with NaN, fill missing names, convert capacity to numeric
gdf = gdf.replace({None: np.nan})
gdf['name'] = gdf['name'].fillna("unbekannt")
gdf['capacity'] = pd.to_numeric(gdf['capacity'], errors='coerce')

# Perform spatial join to associate each daycare center with its district
gdf = gpd.sjoin(
    gdf,
    berlin_districts[['geometry', 'namgem']], 
    how="left",
    predicate="within"
)

# Update district information and clean up redundant columns
gdf['suburb'] = gdf['namgem']
gdf = gdf.drop(columns=['namgem', 'index_right'])


# Create visualization of daycare centers on district map
fig, ax = plt.subplots(figsize=(12, 8))
berlin_districts.plot(ax=ax, alpha=0.5)
gdf.plot(ax=ax, color='red', markersize=1)
plt.title('Kita-Standorte in Berlin')
plt.axis('off')
plt.show()
```

## Analyse der Kita-Geometrien und Kapazitäten

Die OpenStreetMap-Daten enthalten zwei verschiedene Arten von Kita-Geometrien, die getrennt analysiert werden müssen:

1. **Punktdaten (Nodes)**:
Diese repräsentieren Kitas, die einfach als Standortpunkte erfasst wurden. Sie stellen die Mehrheit der Einträge dar und sind typisch für kleinere Einrichtungen.

2. **Flächendaten (Polygons)**:
Diese zeigen Kitas, bei denen die tatsächlichen Gebäudegrundrisse oder Grundstücksflächen kartiert wurden. Dies ist oft bei größeren Einrichtungen der Fall. Für die weitere Analyse werden diese Flächen in Mittelpunkte (Zentroide) umgewandelt.

Die Verteilung der bekannten Kapazitäten wird in einem Histogramm dargestellt. Diese Visualisierung gibt wichtige Einblicke in:
- Die typischen Größenordnungen von Berliner Kitas
- Mögliche Ausreißer oder ungewöhnliche Werte
- Die Bandbreite der Betreuungskapazitäten

Diese Informationen sind wertvoll für die spätere Abschätzung fehlender Kapazitätswerte und die Bewertung der Versorgungssituation.

```python
# Split data by geometry type into points (nodes) and polygons
nodes = gdf[gdf.geometry.type == 'Point'].copy()
polygons = gdf[gdf.geometry.type == 'Polygon'].copy()

# Calculate area for polygons and convert polygon geometries to centroids
polygons['area'] = polygons.geometry.area.round().astype(float)
polygons['geometry'] = polygons.geometry.centroid

print(f"Nodes: {len(nodes)}, Polygons: {len(polygons)}")

# Create a combined dataset of known capacities from both nodes and polygons
known_capacities = pd.concat([
    nodes[nodes['capacity'].notna()]['capacity'],
    polygons[polygons['capacity'].notna()]['capacity']
])

# Visualize the distribution of known capacities using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(known_capacities, bins=30)
plt.title('Distribution of Known Capacities')
plt.xlabel('Capacity')
plt.ylabel('Count')
plt.show()
```

## Kapazitätsschätzung basierend auf Gebäudeflächen

Für Kitas, die als Flächen kartiert sind aber keine Kapazitätsangaben haben, entwickeln wir ein statistisches Modell zur Schätzung der Betreuungsplätze. Dieser Ansatz basiert auf der Annahme, dass die Gebäudefläche einer Kita in einem systematischen Zusammenhang mit ihrer Betreuungskapazität steht.

Die Analyse erfolgt in mehreren Schritten:

1. **Korrelationsanalyse**:
Ein Streudiagramm stellt den Zusammenhang zwischen Gebäudefläche und bekannten Kapazitäten dar. Dies gibt einen ersten visuellen Eindruck der Beziehung dieser Variablen.

2. **Regressionsmodell**:
Mit Hilfe einer linearen Regression wird der mathematische Zusammenhang zwischen Fläche und Kapazität modelliert. Die Regressionsstatistiken geben Aufschluss über die Güte des Modells und die Verlässlichkeit der Schätzungen.

3. **Kapazitätsvorhersage**:
Für Kitas ohne Kapazitätsangaben werden basierend auf ihrer Fläche Betreuungsplätze geschätzt. Dabei werden auch Konfidenzintervalle berechnet, die die Unsicherheit der Schätzungen quantifizieren.

4. **Ausreißerbehandlung**:
Ungewöhnlich hohe Schätzwerte werden durch statistische Methoden (IQR-Methode) identifiziert und durch realistischere Werte ersetzt. Dies verhindert unrealistische Überschätzungen der Kapazitäten.

Die abschließende Visualisierung zeigt die geschätzten Kapazitäten, wobei potenzielle Ausreißer rot markiert sind. Diese bereinigten Schätzwerte vervollständigen unseren Datensatz für die weitere Analyse der Versorgungssituation.

```python
# Filter data to only include polygons with valid capacity and area values
valid_data = polygons[polygons['capacity'].notna() & polygons['area'].notna()]

# Create scatter plot to visualize relationship between area and capacity
plt.figure(figsize=(10, 6))
plt.scatter(valid_data['area'], valid_data['capacity'])
plt.xlabel('Area (m²)') 
plt.ylabel('Capacity')
plt.title('Relationship between Area and Capacity')
plt.show()

# Perform linear regression analysis
# Reshape area values and add constant term for regression
X = valid_data['area'].values.reshape(-1, 1)
X = sm.add_constant(X)
y = valid_data['capacity'].values
# Fit ordinary least squares regression model
model = sm.OLS(y, X).fit()

print("\nRegression Statistics:")
print(model.summary().tables[1])
```

#### Interpretation der Regressionsanalyse

Die Regressionsanalyse zeigt einen signifikanten Zusammenhang zwischen der Gebäudefläche und der Kita-Kapazität:

1. **Grundkapazität** (const = 68.59):
- Eine Kita hat im Durchschnitt eine "Basis-Kapazität" von etwa 69 Plätzen, unabhängig von ihrer Größe
- Der Wert ist statistisch hochsignifikant (p < 0.001)

2. **Flächeneffekt** (x1 = 0.0132):
- Pro zusätzlichem Quadratmeter Fläche kommen durchschnittlich 0,0132 Betreuungsplätze hinzu
- Oder anders ausgedrückt: Pro 100m² zusätzlicher Fläche gibt es etwa 1,32 zusätzliche Betreuungsplätze
- Auch dieser Zusammenhang ist statistisch hochsignifikant (p < 0.001)

Die Konfidenzintervalle ([0.025 0.975]) zeigen, dass die wahren Werte mit 95% Wahrscheinlichkeit für die Grundkapazität zwischen 55 und 82 Plätzen und für den Flächeneffekt zwischen 0,009 und 0,017 Plätzen pro m² liegen.

```python
# Predict capacity for polygons with missing values
missing_capacity = polygons[polygons['capacity'].isna()].copy()
X_pred = sm.add_constant(missing_capacity['area'].values.reshape(-1, 1))
missing_capacity['predicted_capacity'] = model.predict(X_pred)

# Transfer predictions back to main polygons DataFrame
polygons.loc[polygons['capacity'].isna(), 'predicted_capacity'] = missing_capacity['predicted_capacity']

# Calculate confidence intervals for predictions
prediction = model.get_prediction(X_pred)
missing_capacity['conf_lower'], missing_capacity['conf_upper'] = prediction.conf_int().T

# Identify outliers using IQR method
Q1 = missing_capacity['predicted_capacity'].quantile(0.25)
Q3 = missing_capacity['predicted_capacity'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
missing_capacity['is_outlier'] = missing_capacity['predicted_capacity'] > upper_bound

# Replace outliers with the median of the non-outliers
median_value = missing_capacity[~missing_capacity['is_outlier']]['predicted_capacity'].median()
missing_capacity.loc[missing_capacity['is_outlier'], 'predicted_capacity'] = median_value


# Visualize predictions highlighting outliers
plt.figure(figsize=(10, 6))
plt.scatter(missing_capacity[~missing_capacity['is_outlier']]['area'], missing_capacity[~missing_capacity['is_outlier']]['predicted_capacity'],label='Normal')
plt.scatter(missing_capacity[missing_capacity['is_outlier']]['area'],missing_capacity[missing_capacity['is_outlier']]['predicted_capacity'],color='red', label='Outliers')
plt.xlabel('Area (m²)')
plt.ylabel('Predicted Capacity')
plt.title('Predicted Capacities with Outliers')
plt.legend()
plt.show()
```

## Bezirksspezifische Kapazitätsschätzung für Punkt-Kitas

Für Kitas, die nur als Punkte erfasst sind und keine Kapazitätsangaben haben, wird eine bezirksspezifische Schätzmethode entwickelt. Diese berücksichtigt lokale Unterschiede in den Kita-Größen:

1. **Berechnung der Referenzwerte**:
- Für jeden Bezirk wird der Median der bekannten Kapazitäten ermittelt
- Zusätzlich wird ein berlinweiter Median als Ausweichlösung berechnet
- Diese Werte dienen als Ausgangsbasis für die Schätzungen

2. **Schätzmethodik**:
- Für fehlende Werte wird der jeweilige Bezirksmedian als Grundlage verwendet
- Falls für einen Bezirk keine Referenzwerte vorliegen, wird der Berliner Gesamtmedian genutzt
- Eine zufällige Variation von ±15% wird hinzugefügt, um realistische Streuung zu erzeugen
- Die Werte werden auf einen plausiblen Bereich von 10 bis 200 Plätzen begrenzt

3. **Visualisierung der Ergebnisse**:
- Ein Boxplot zeigt die Verteilung der geschätzten Kapazitäten nach Bezirken
- Ein gestapeltes Histogramm visualisiert die Gesamtverteilung der Schätzwerte
- Ausreißer werden gesondert dargestellt

Diese lokalisierte Schätzmethode berücksichtigt bezirksspezifische Unterschiede in der Kita-Größenstruktur und erzeugt dadurch realistischere Kapazitätsschätzungen als ein berlinweiter Einheitswert.

```python
# Calculate district-wise medians
known_capacities = pd.concat([
    nodes[nodes['capacity'].notna()][['capacity', 'suburb']],
    polygons[['capacity', 'suburb']]
])
global_median = known_capacities['capacity'].median()  # Calculate the global median
district_medians = known_capacities.groupby('suburb')['capacity'].median()  # Calculate district-wise medians

# Estimate missing capacities
nodes_missing = nodes[nodes['capacity'].isna()].copy()
nodes_missing['predicted_capacity'] = np.nan  # Initialize a column for predicted capacities

for suburb in nodes_missing['suburb'].unique():
    mask = nodes_missing['suburb'] == suburb
    n_samples = mask.sum()  # Count the number of samples for the current suburb
    
    # Determine the base capacity for the current suburb
    base_capacity = district_medians.get(suburb, global_median)
    variation = np.random.uniform(0.85, 1.15, n_samples)  # Generate random variation
    capacities = (base_capacity * variation).round()  # Calculate capacities with variation
    capacities = np.clip(capacities, 10, 200)  # Clip capacities within a reasonable range
    
    nodes_missing.loc[mask, 'predicted_capacity'] = capacities  # Assign predicted capacities

nodes.loc[nodes['capacity'].isna(), 'predicted_capacity'] = nodes_missing['predicted_capacity']  # Transfer predictions back to the main DataFrame

# Visualization of estimated capacities
plt.figure(figsize=(10, 6))
sns.boxplot(x='suburb', y='predicted_capacity', data=nodes_missing)  # Boxplot for visualizing estimated capacities by suburb
plt.xticks(rotation=90)
plt.title('Estimated Capacities by District')
plt.xlabel('District')
plt.ylabel('Estimated Capacity')
plt.show()

# Distribution of predicted capacities
plt.figure(figsize=(10, 6))
sns.histplot(data=missing_capacity, x='predicted_capacity', hue='is_outlier', multiple="stack")  # Histogram for visualizing the distribution of predicted capacities
plt.title('Distribution of Predicted Capacities')
plt.show()
```

## Skalierung der geschätzten Kapazitäten auf Zielwert

Um sicherzustellen, dass unsere Kapazitätsschätzungen mit der bekannten Gesamtzahl der Betreuungsplätze in Berlin übereinstimmen, wird eine finale Anpassung vorgenommen:

1. **Berechnung der Gesamtkapazität**:
- Summierung aller bekannten und geschätzten Kapazitäten aus Punkt- und Flächendaten
- Diese erste Summe basiert auf unseren statistischen Schätzungen

2. **Skalierungsfaktor**:
- Vergleich der geschätzten Gesamtkapazität mit der offiziellen Zielkapazität
- Berechnung eines Skalierungsfaktors, der die Schätzwerte auf den Zielwert adjustiert

3. **Anpassung der Schätzungen**:
- Alle geschätzten (nicht die bekannten!) Kapazitätswerte werden mit dem Skalierungsfaktor multipliziert
- Dies stellt sicher, dass die Gesamtsumme der Betreuungsplätze mit der Realität übereinstimmt
- Die relative Verteilung der geschätzten Kapazitäten bleibt dabei erhalten

Diese Skalierung ist der letzte Schritt der Kapazitätsschätzung und gewährleistet die Konsistenz unserer Daten mit den offiziellen Statistiken.

```python
# Calculate the current total capacity
total_capacity = (
    nodes['capacity'].fillna(nodes['predicted_capacity']).sum() +
    polygons['capacity'].fillna(polygons['predicted_capacity']).sum()
)

# Adjust to target capacity
scale_factor = TARGET_CAPACITY / total_capacity
print(f"Scale factor: {scale_factor:.3f}")

# Scale the estimated capacities
for gdf in [nodes, polygons]:
    mask = gdf['capacity'].isna()
    gdf.loc[mask, 'predicted_capacity'] = (
        gdf.loc[mask, 'predicted_capacity'] * scale_factor
    ).round()
```

## Finalisierung und Analyse des Kapazitätsdatensatzes

In diesem abschließenden Schritt werden die Daten zusammengeführt, analysiert und für die weitere Verwendung aufbereitet:

1. **Datenzusammenführung**:
- Vereinigung der Punkt- und Flächendaten zu einem Gesamtdatensatz
- Erstellung einer einheitlichen Kapazitätsspalte mit beobachteten und geschätzten Werten
- Dokumentation der Datenherkunft (beobachtet/geschätzt) für Transparenz

2. **Statistische Auswertung**:
- Gesamtübersicht der Berliner Kita-Landschaft (Anzahl, Gesamtkapazität)
- Detaillierte Bezirksstatistiken mit:
 - Summe der Betreuungsplätze
 - Durchschnittliche Kita-Größe
 - Anzahl der Einrichtungen

3. **Visualisierungen**:
- Boxplots zur Verteilung der Kapazitäten nach Datenquelle
- Balkengrafik der Gesamtkapazitäten nach Bezirken
- Gestapeltes Histogramm zur Visualisierung der Kapazitätsverteilung

4. **Datenspeicherung**:
- Erstellung eines finalen GeoJSON-Datensatzes mit allen relevanten Informationen
- Sicherung der Daten für weitere Analysen und Visualisierungen

Diese finale Aufbereitung bildet die Grundlage für alle weiteren Analysen im Rahmen des KitaVision-Projekts und ermöglicht tiefgehende Einblicke in die Versorgungssituation.

```python
# Selection and merging of relevant columns
columns = ['name', 'id', 'capacity', 'predicted_capacity', 'geometry', 'suburb']
combined = pd.concat([nodes[columns], polygons[columns]], ignore_index=True)

# Creation of the final capacity columns
combined['final_capacity'] = combined['capacity'].fillna(combined['predicted_capacity'])
combined['capacity_source'] = np.where(
    combined['capacity'].isna(),
    'predicted',
    'observed'
)
```

```python
# Print final statistics
print("\nFinale Statistiken:")
print(f"Gesamtanzahl Kitas: {len(combined):,}")
print(f"Geschätzte Kapazitäten: {(combined['capacity_source'] == 'predicted').sum():,}")
print(f"Gesamtkapazität: {combined['final_capacity'].sum():,.0f}")

# Calculate capacity distribution statistics by district
# Aggregates sum, mean and count of final capacity for each suburb
bezirk_stats = combined.groupby('suburb').agg({
    'final_capacity': ['sum', 'mean', 'count']
}).round(0)
display(bezirk_stats)
```

```python
# Select final columns for output dataset
# - name: Name of the daycare center
# - id: Unique identifier
# - suburb: District/neighborhood
# - final_capacity: Final capacity (either observed or predicted)
# - capacity_source: Source of capacity data (observed/predicted)
# - geometry: Geographic location/shape
final_columns = ['name', 'id', 'suburb', 'final_capacity', 'capacity_source', 'geometry']
combined = combined[final_columns]

# Create output directory if it doesn't exist
OUTPUT_FILE.parent.mkdir(parents=True, exist_ok=True)

# Save final dataset as GeoJSON file
combined.to_file(OUTPUT_FILE, driver='GeoJSON')
print(f"\nData saved to: {OUTPUT_FILE}")
```

```python
# Create visualization layout with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Create boxplot showing capacity distribution by data source (observed vs predicted)
sns.boxplot(x='capacity_source', y='final_capacity', data=combined, ax=ax1)
ax1.set_title('Capacity Distribution by Source')

# Calculate and plot total capacity per district as bar chart
bezirk_stats = combined.groupby('suburb')['final_capacity'].sum().sort_values()
bezirk_stats.plot(kind='bar', ax=ax2)
ax2.set_title('Total Capacity per District')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Additional analysis: Create stacked histogram showing capacity distribution by source
plt.figure(figsize=(10, 6))
sns.histplot(data=combined, x='final_capacity', hue='capacity_source', multiple="stack")
plt.title('Distribution of Capacities by Data Source')
plt.show()
```
