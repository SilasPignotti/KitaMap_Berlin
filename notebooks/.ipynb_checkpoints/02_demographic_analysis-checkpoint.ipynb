# Bevölkerungsprognose für Kinder im Alter von 0–6 Jahren in Berlin

Dieses Notebook ist Teil des Projekts **„KitaVision Berlin“** und zeigt, wie die historischen Kinderzahlen pro Bezirk aufbereitet, analysiert und mittels verschiedener Zeitreihenmodelle fortgeschrieben werden. Ziel ist es, belastbare Prognosen zur Entwicklung der Kinderzahlen (0–6 Jahre) bis 2039 abzuleiten. Diese Daten fließen in die Gesamtschau zur Versorgungssituation mit Kita-Plätzen ein und bilden damit eine wichtige Grundlage für bedarfsgerechte Planungen und Entscheidungen.

## Kernaufgaben des Notebooks:

1. **Dateneinbindung und Historienanalyse**  
   - Einlesen und Filtern der Daten zu Kinderzahlen (Jahre 2015–2024)  
   - Erste visuelle Analyse zur Veranschaulichung der Trends in den Bezirken

2. **Modellauswahl und Prognoseerstellung**  
   - Anwendung verschiedener Vorhersagemodelle (Exponential Smoothing, Prophet)  
   - Entwicklung eines kombinierten Ensemble-Modells zur Verbesserung der Genauigkeit

3. **Modellbewertung und Ergebnisaufbereitung**  
   - Berechnung von Fehlerkennzahlen (RMSE, MAE, MAPE) zur objektiven Modellbeurteilung  
   - Übertragung der finalen Prognosewerte in einen strukturierten Datensatz und CSV-Export  
   - Visualisierung, um historische und prognostizierte Entwicklungen direkt vergleichen zu können

Diese Prognoseergebnisse unterstützen die Planung von Kita-Kapazitäten, indem sie frühzeitig auf potenzielle Engpässe oder steigende Nachfrage in bestimmten Bezirken hinweisen. Somit leistet das Notebook einen wesentlichen Beitrag zur vorausschauenden Gestaltung der Berliner Kita-Landschaft.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')  # Unterdrückt Warnungen für übersichtlichere Ausgabe

# Define data paths
INPUT_PATH = Path("../data/raw/entwicklung_2015_2024.csv")
OUTPUT_PATH = Path("../data/processed/prognose_2024_2054.csv")
```

## Entwicklung der Kinderzahlen (2015–2024) und Grundgerüst für Prognosen

In diesem Abschnitt werden die historisch erfassten Kinderzahlen (0–6 Jahre) je Berliner Bezirk aus einer CSV-Datei eingelesen, visuell aufbereitet und für weiterführende Prognosen vorbereitet:

1. **Dateneinlesung**  
   Die CSV-Daten werden gefiltert, sodass nur der Bezirksname und die Jahreswerte von 2015 bis 2024 erhalten bleiben.

2. **Visualisierung**  
   Ein Liniendiagramm zeigt die zeitliche Entwicklung der Kinderzahlen pro Bezirk. Durch unterschiedliche Farblinien und eine externe Legende entsteht ein übersichtlicher Vergleich zwischen den Bezirken.

3. **Prognose-Grundgerüst**  
   Für die zukünftigen Jahre (2024, 2029, 2034, 2039) wird ein neuer DataFrame erstellt. Die Werte für 2024 werden als Ausgangsbasis direkt aus den historischen Daten übernommen. Dadurch entsteht ein erster Rahmen für die spätere Fortschreibung der Kinderzahlen.


```python
# Load historical data
df = pd.read_csv(INPUT_PATH)
year_columns = [str(year) for year in range(2015, 2025)]  # Years from 2015-2024
df = df[['Bezirk'] + year_columns]  # District + years in chronological order

# Plotting Setup
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Create a nice color palette for the districts
colors = sns.color_palette("husl", n_colors=len(df))

# Plot for each district
for idx, row in df.iterrows():
    district = row['Bezirk']
    values = row[year_columns].values  # Now we can directly take the values
    plt.plot(year_columns, values, marker='o', label=district, color=colors[idx], linewidth=2)

# Plot Layout
plt.title('Development of Child Numbers 2015-2024', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Children', fontsize=12)
plt.xticks(rotation=45)

# Add thousand separators to the Y-axis
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), 
            loc='upper left', 
            borderaxespad=0.,
            title='Districts',
            frameon=True)

plt.tight_layout()
plt.show()

# Create results DataFrame with districts as index and forecast years as columns
forecast_years = [2024, 2029, 2034, 2039]
results = pd.DataFrame(index=df['Bezirk'], columns=forecast_years)

# Fill in 2024 values
results[2024] = df['2024'].values

print("\nResults DataFrame Structure:")
print(results)
```

## Prognosefehler-Metriken und Modellbewertung

In diesem Abschnitt werden zunächst drei Funktionen definiert, um Prognosefehler zu quantifizieren. Diese Kennzahlen ermöglichen eine objektive Bewertung unterschiedlicher Vorhersagemodelle:

1. **RMSE (Root Mean Squared Error)**  
   Liefert die mittlere Abweichung der Prognosen im Quadrat, indem es die Wurzel aus dem Mittel der quadrierten Fehler zieht.

2. **MAE (Mean Absolute Error)**  
   Gibt an, wie hoch die durchschnittliche Abweichung der Prognosen in absoluten Werten ist.

3. **MAPE (Mean Absolute Percentage Error)**  
   Misst die prozentuale Abweichung zwischen vorhergesagten und tatsächlichen Werten und zeigt damit an, wie viel Prozent das Modell im Durchschnitt danebenliegt.

Zusätzlich werden die Zieljahre für die Prognosen (2029, 2034, 2039) definiert und in einem neuen DataFrame **all_predictions** festgehalten. Dieser DataFrame dient als strukturierte Übersicht für die Vorhersageergebnisse der drei Modelle **ETS**, **Prophet** und **Ensemble**.  

Abschließend wird eine Funktion **evaluate_model** definiert, die zu einem bestimmten Bezirk die vorhergesagten Werte eines Modells mit den tatsächlichen Werten vergleicht und die oben genannten Fehlermaße berechnet. Die Ergebnisse werden in einem Dictionary zusammengefasst und können anschließend für Vergleiche und Auswertungen herangezogen werden.


```python
def calculate_rmse(y_true, y_pred):
    # Calculate the root mean squared error
    return sqrt(mean_squared_error(y_true, y_pred))

def calculate_mae(y_true, y_pred):
    # Calculate the mean absolute error
    return mean_absolute_error(y_true, y_pred)

def calculate_mape(y_true, y_pred):
    # Calculate the mean absolute percentage error
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate_model(model_name, y_true, y_pred, bezirk):
    # Evaluate the model performance
    return {
        'Model': model_name,
        'Bezirk': bezirk,
        'RMSE': calculate_rmse(y_true, y_pred),
        'MAE': calculate_mae(y_true, y_pred),
        'MAPE': calculate_mape(y_true, y_pred)
    }
```

```python
# Define forecast years
forecast_years = [2029, 2034, 2039]

# Initialize DataFrame for all predictions
all_predictions = pd.DataFrame(
    index=df['Bezirk'],
    columns=[
        f'{model}_{year}' 
        for model in ['ETS', 'Prophet', 'Ensemble'] 
        for year in forecast_years
    ]
)

evaluation_results = []
```

## Exponential Smoothing-Vorhersage und Modellbewertung

In diesem Abschnitt wird das Exponential Smoothing-Verfahren auf die Kinderzahlen aller Berliner Bezirke angewendet. Für jeden Bezirk erfolgt zunächst eine Aufteilung in Trainingsdaten (alle bis auf die letzten zwei Werte) und Testdaten (die letzten zwei Werte). Das Modell wird dann trainiert und erstellt eine Vorhersage für 16 zukünftige Zeitpunkte, aus denen anschließend die Werte für die festgelegten Prognosejahre (z. B. 2029, 2034, 2039) entnommen werden.  

Zur Bewertung wird ein separates Modell mit den Trainingsdaten erneut angepasst und auf die Testdaten angewendet. Die Prognosegüte wird anhand der zuvor definierten Kennzahlen **RMSE**, **MAE** und **MAPE** ermittelt, um einen objektiven Vergleich zwischen vorhergesagten und tatsächlichen Werten zu ermöglichen. Die Ergebnisse werden zur Übersicht in einem DataFrame zusammengefasst und nach dem Mittelwert aller Fehlermaße gruppiert.


```python
# Exponential Smoothing for each district
for district in df['District']:
    # Prepare historical data - now in correct chronological order
    historical_values = df[df['District'] == district][year_columns].values.flatten()
    
    # Train-Test-Split for evaluation
    train_data = historical_values[:-2]
    test_data = historical_values[-2:]
    
    # Exponential Smoothing forecast
    exp_model = ExponentialSmoothing(
        historical_values, trend='add', seasonal=None, damped_trend=True
    ).fit()
    exp_forecast = exp_model.forecast(16)
    
    # Save the ETS forecasts
    for i, year in enumerate(forecast_years):
        all_predictions.loc[district, f'ETS_{year}'] = np.round(exp_forecast[4 + i*5], 0)
    
    # Evaluation
    exp_model_eval = ExponentialSmoothing(
        train_data, trend='add', seasonal=None, damped_trend=True
    ).fit()
    exp_pred_eval = exp_model_eval.forecast(2)
    evaluation_results.append(
        evaluate_model('Exponential Smoothing', test_data, exp_pred_eval, district)
    )

# Display results for ETS
print("\nExponential Smoothing forecasts:")
print(all_predictions.filter(like='ETS'))
print("\nETS Evaluation:")
ets_eval = pd.DataFrame(evaluation_results)
print(ets_eval.groupby('Model')[['RMSE', 'MAE', 'MAPE']].mean())
```

## Prophet-Vorhersage und Modellbewertung

Auch für das **Prophet**-Modell wird zunächst die zeitliche Entwicklung der Kinderzahlen pro Bezirk herangezogen. Ausgehend von den historischen Werten (2015–2024) werden entsprechende Zeitstempel erstellt, die Prophet als Eingabe dienen. Danach wird das Modell trainiert und für die definierten Vorhersagejahre (etwa 2029, 2034 und 2039) eine Prognose erzeugt.

Für die Evaluierung wird das Modell auf einen verkürzten Datensatz (ohne die letzten beiden Werte) fitten und die Performance anhand der verbleibenden zwei Testpunkte messen. Die vorhergesagten Werte werden mit den tatsächlichen verglichen und durch **RMSE**, **MAE** und **MAPE** quantifiziert. Abschließend fasst ein DataFrame die Fehlermetriken je Bezirk zusammen, was einen direkten Vergleich der Modellgenauigkeit ermöglicht.


```python
# Prophet for each district
for district in df['District']:
    # Extract historical data - and flatten() to make it 1-dimensional
    historical_values = df[df['District'] == district][year_columns].values.flatten()
    
    # Create date range
    dates = pd.date_range(start='2015', periods=len(year_columns), freq='Y')
    
    # Prepare data for Prophet
    prophet_df = pd.DataFrame({
        'ds': dates,
        'y': historical_values
    })
    
    # Fit model and make forecast
    prophet_model = Prophet(yearly_seasonality=False)
    prophet_model.fit(prophet_df)
    
    # Create future data
    future = pd.DataFrame({
        'ds': [pd.Timestamp(str(year)) for year in forecast_years]
    })
    prophet_forecast = prophet_model.predict(future)
    
    # Extract forecasts for the desired years
    for year in forecast_years:
        forecast_value = prophet_forecast[prophet_forecast['ds'].dt.year == year]['yhat'].values[0]
        all_predictions.loc[district, f'Prophet_{year}'] = np.round(forecast_value, 0)
    
    # Evaluation
    train_data = historical_values[:-2]
    test_data = historical_values[-2:]
    
    # Prepare evaluation dataset
    train_dates = dates[:-2]
    prophet_df_eval = pd.DataFrame({
        'ds': train_dates,
        'y': train_data
    })
    
    # Fit evaluation model
    prophet_model_eval = Prophet(yearly_seasonality=False)
    prophet_model_eval.fit(prophet_df_eval)
    
    # Forecast for evaluation period
    future_eval = pd.DataFrame({'ds': dates[-2:]})
    prophet_pred_eval = prophet_model_eval.predict(future_eval)['yhat'].values
    
    evaluation_results.append(
        evaluate_model('Prophet', test_data, prophet_pred_eval, district)
    )

# Display results for Prophet
print("\nProphet forecasts:")
print(all_predictions.filter(like='Prophet'))
print("\nProphet Evaluation:")
prophet_eval = pd.DataFrame(evaluation_results)
print(prophet_eval[prophet_eval['Model'] == 'Prophet'].groupby('Model')[['RMSE', 'MAE', 'MAPE']].mean())
```

## Ensemble-Vorhersage und finale Modellkombination

In diesem Abschnitt werden die Ergebnisse der beiden Modellansätze **Exponential Smoothing (ETS)** und **Prophet** zu einer gewichteten Ensemblevorhersage kombiniert. Hierfür dient der mittlere absolute prozentuale Fehler (**MAPE**) als Maß für die Güte der einzelnen Modelle:

1. **Modellgewichtung**  
   - Die MAPE beider Modelle wird berechnet; Modelle mit niedrigerem MAPE erhalten ein höheres Gewicht.  
   - Das Gewicht ergibt sich aus dem inversen Verhältnis des jeweiligen MAPE-Werts.

2. **Ensembleprognose je Bezirk**  
   - Für jedes Prognosejahr wird eine gewichtete Mittelung der beiden Vorhersagen (ETS und Prophet) durchgeführt.  
   - So entsteht eine kombinierte Vorhersage, die die Stärken beider Methoden vereint.

3. **Ensemble-Evaluierung**  
   - Zur Bewertung wird anhand der letzten beiden Datenpunkte ein erneuter Test durchgeführt.  
   - Anhand von **RMSE**, **MAE** und **MAPE** lässt sich anschließend überprüfen, ob das Ensemble einen Mehrwert gegenüber den Einzelmodellen liefert.

Abschließend werden die finalen Ensemble-Prognosen in den **results**-DataFrame übertragen. Dies bildet die Grundlage für weiterführende Analysen und Bewertungen der zukünftigen Entwicklung der Kinderzahlen.


```python
# Method 3: Ensemble

# Calculate weights based on the performance (MAPE) of the models
ets_mape = ets_eval[ets_eval['Model'] == 'Exponential Smoothing']['MAPE'].mean()
prophet_mape = prophet_eval[prophet_eval['Model'] == 'Prophet']['MAPE'].mean()

# Use inverse MAPE for weighting
ets_weight = (1/ets_mape) / (1/ets_mape + 1/prophet_mape)
prophet_weight = (1/prophet_mape) / (1/ets_mape + 1/prophet_mape)

# Ensemble for each district
for district in df['District']:
    # Extract historical data in correct order
    historical_values = df[df['District'] == district][year_columns].values.flatten()
    dates = pd.date_range(start='2015', periods=len(year_columns), freq='Y')
    
    # Calculate ensemble forecast and save
    for year in forecast_years:
        ets_pred = all_predictions.loc[district, f'ETS_{year}']
        prophet_pred = all_predictions.loc[district, f'Prophet_{year}']
        ensemble_pred = (ets_weight * ets_pred + prophet_weight * prophet_pred)
        all_predictions.loc[district, f'Ensemble_{year}'] = np.round(ensemble_pred, 0)
    
    # Evaluation
    train_data = historical_values[:-2]
    test_data = historical_values[-2:]
    
    # ETS Evaluation
    exp_model_eval = ExponentialSmoothing(
        train_data, trend='add', seasonal=None, damped_trend=True
    ).fit()
    ets_pred_eval = exp_model_eval.forecast(2)
    
    # Prophet Evaluation
    train_dates = dates[:-2]
    prophet_df_eval = pd.DataFrame({
        'ds': train_dates,
        'y': train_data
    })
    prophet_model_eval = Prophet(yearly_seasonality=False)
    prophet_model_eval.fit(prophet_df_eval)
    future_eval = pd.DataFrame({'ds': dates[-2:]})
    prophet_pred_eval = prophet_model_eval.predict(future_eval)['yhat'].values
    
    # Ensemble Evaluation
    ensemble_pred_eval = ets_weight * ets_pred_eval + prophet_weight * prophet_pred_eval
    
    evaluation_results.append(
        evaluate_model('Ensemble', test_data, ensemble_pred_eval, district)
    )

# Display results for Ensemble
print("\nEnsemble forecasts:")
print(all_predictions.filter(like='Ensemble'))
print("\nEnsemble Evaluation:")
ensemble_eval = pd.DataFrame(evaluation_results)
print(ensemble_eval[ensemble_eval['Model'] == 'Ensemble'].groupby('Model')[['RMSE', 'MAE', 'MAPE']].mean())

# Transfer final Ensemble forecasts to results DataFrame
for year in forecast_years:
    if year == 2024:
        continue  # 2024 we already have from real data
    results[year] = all_predictions[f'Ensemble_{year}']
```

## Zusammenfassung der Modellbewertung

In diesem abschließenden Schritt werden alle zuvor berechneten Fehlermetriken in einem DataFrame zusammengeführt. Dadurch erhalten wir einen umfassenden Überblick über die Güte der einzelnen Modelle:

1. **Durchschnittliche Kennzahlen pro Modell**  
   Mithilfe einer gruppierten Auswertung (GroupBy) werden die mittleren Werte für **RMSE**, **MAE** und **MAPE** je Modell ermittelt. Dies gibt Aufschluss darüber, welches Verfahren über alle Bezirke hinweg am verlässlichsten prognostiziert.

2. **Bestes Modell pro Bezirk**  
   Zusätzlich wird für jeden Bezirk das Modell mit dem geringsten MAPE-Wert identifiziert. Dies zeigt, in welchen Bezirken bestimmte Modelle besonders gute Ergebnisse liefern und unterstützt eine bedarfsorientierte Modellauswahl.

Diese Auswertungen bilden die Basis für eine fundierte Entscheidung, ob ein Modell global (für alle Bezirke) oder eher lokal (je nach Bezirk) am besten geeignet ist.


```python
# Convert Evaluation Results to DataFrame
eval_df = pd.DataFrame(evaluation_results)

# Average metrics per model
print("Average metrics per model:")
print(eval_df.groupby('Model')[['RMSE', 'MAE', 'MAPE']].mean())

# Best models per district based on MAPE
print("\nBest models per district (based on MAPE):")
best_models = eval_df.loc[eval_df.groupby('Bezirk')['MAPE'].idxmin()]
print(best_models[['Bezirk', 'Model', 'MAPE']])
```

## Finalisierung der Prognosen und Visualisierung

In diesem abschließenden Schritt werden die ermittelten Prognosewerte auf ganze Zahlen gerundet und in einer CSV-Datei gespeichert. Anschließend erfolgt eine kombinierte Darstellung von historischen Daten (2015–2024) und den prognostizierten Werten (2024, 2029, 2034, 2039) in einem Liniendiagramm:

1. **Datenspeicherung**  
   - Die gerundeten Werte werden in eine CSV-Datei exportiert, um sie außerhalb der Notebook-Umgebung weiterverarbeiten zu können.

2. **Darstellung der Zeitreihen**  
   - Die historischen Werte jedes Bezirks werden als durchgezogene Linien aufgezeigt, sodass ein Überblick über die bisherige Entwicklung entsteht.  
   - Die Prognosewerte werden im selben Diagramm als gestrichelte Linien dargestellt, um die zu erwartende Entwicklung bis 2039 zu verdeutlichen.  
   - Durch unterschiedliche Farbcodierungen kann man die zeitlichen Verläufe der Bezirke leicht vergleichen.

3. **Plot-Layout**  
   - Titel, Achsenbeschriftungen und eine Legende neben dem Diagramm sorgen für eine übersichtliche Darstellung.  
   - Der Gitterhintergrund erleichtert die Ablesbarkeit der Werte, während der Y-Achsenbereich so eingestellt ist, dass sämtliche Bezirke gut sichtbar sind.

Diese finale Visualisierung ermöglicht eine intuitive Einschätzung der bisherigen und zukünftigen Entwicklung der Kinderzahlen in Berlin.


```python
# Round results to whole numbers and save
results = results.astype(int)

# Save results as CSV
results.to_csv(OUTPUT_PATH)

# Output the save path
print(f"\nForecast saved in: {OUTPUT_PATH}")

# Visualization of the final forecasts
plt.figure(figsize=(15, 8))

# Define colors for each district
colors = plt.cm.tab20(np.linspace(0, 1, len(df['Bezirk'])))

# Plot historical data and forecasts for each district
for idx, bezirk in enumerate(df['Bezirk']):
        # Historical data
        historical_values = df[df['Bezirk'] == bezirk].iloc[0, 1:].values
        historical_years = list(range(2015, 2025))
        
        # Forecast values
        forecast_values = results.loc[bezirk]
        all_years = results.columns.astype(int)
        
        # Plot with the same color for historical and forecast
        plt.plot(historical_years, historical_values, 
                marker='o', 
                linestyle='-', 
                color=colors[idx],
                linewidth=2,
                label=f'{bezirk} (Historical)')
        
        plt.plot(all_years, forecast_values, 
                marker='s', 
                linestyle='--', 
                color=colors[idx],
                linewidth=2,
                label=f'{bezirk} (Forecast)')

plt.title('Development and Forecast of Child Numbers 2015-2039', pad=20, size=14)
plt.xlabel('Year', size=12)
plt.ylabel('Number of Children', size=12)

# Adjust grid
plt.grid(True, linestyle='--', alpha=0.7, color='gray', zorder=0)

# Adjust axis range
plt.ylim(bottom=12000)  # Adjust lower bound

# Optimize legend
plt.legend(bbox_to_anchor=(1.05, 1), 
        loc='upper left', 
        borderaxespad=0,
        frameon=True,
        fancybox=True,
        shadow=True,
        fontsize=10)

# Adjust layout
plt.tight_layout()
plt.show()
```
